```python
import numpy as np
from scipy.linalg import solve

def mean_recurrence_time(Q):
    n = len(Q)
    mean_recurrence_times = []

    for state in range(n):
        # Remove the row and column corresponding to the specified state
        A = np.delete(Q, state, axis=0)
        A = np.delete(A, state, axis=1)

        # Construct the vector b filled with -1's
        b = -np.ones(n - 1)

        # Construct the vector c to remove other rows that are not in the specified state
        c = Q[state]
        # Filter out negative entries
        c_positive = [x for x in c if x >= 0]

        # Solve the system of linear equations A * m = b
        m = solve(A, b)

        # Calculate the mean sojourn time
        sojourn_time = np.sum(Q[state][Q[state] > 0])

        # Calclulate the mean recurrence time [IT'S ONE OF THESE FORMULAS BUT THE TEXTBOOK AND LAWLEY GAVE DIFFERENT FORMULAS]
        #MRT = np.sum((1 + np.dot(m, c_positive)) / sojourn_time) # FROM LAWLEY
        MRT = np.sum(((n-1) + np.dot(m, c_positive)) / sojourn_time) # FROM THE TEXTBOOK
        
        mean_recurrence_times.append(MRT)

    return mean_recurrence_times

# Q = np.array([[-3, 1, 2],
#              [3, -3.5, 0.5],
#              [0, 1, -1]])

Q = np.array([[-5/6, 1/3, 1/2, 0],
              [0, -1/3, 0, 1/3],
              [0, 0, -1/2, 1/2],
              [1/6, 0, 0, -1/6]])

mean_recurrence_times = mean_recurrence_time(Q)
for state, MRT in enumerate(mean_recurrence_times):
    print("Mean recurrence time for state", state + 1, ":", MRT)
```
